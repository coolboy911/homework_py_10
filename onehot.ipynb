# Решение

````python
import random
import pandas as pd
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head(10)
````

не очень компактное решение в лоб

````python
data.loc[data['whoAmI'] == "robot", 'is_robot'] = 1
data.loc[data['whoAmI'] != "robot", 'is_robot'] = 0
data.loc[data['whoAmI'] == "human", 'is_human'] = 1
data.loc[data['whoAmI'] != "human", 'is_human'] = 0
````

````python
data.head(10)
````

код снизу мне подсказал chat gpt бот в телеграме

````python
data['is_robot'] = (data['whoAmI'] == 'robot').astype(int)
data['is_human'] = (data['whoAmI'] == 'human').astype(int)
````
